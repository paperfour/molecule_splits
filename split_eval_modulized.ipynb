```python
# Testing Karger's algorithm 

import numpy as np
from karger_alg import generate_karger

# TODO: See function
from cross_split import tanimoto_cso




# Create a dummy adjacency matrix
dummy_size = 50
adj_mat = np.random.rand(dummy_size, dummy_size)
np.fill_diagonal(adj_mat, 0)
adj_mat = adj_mat + adj_mat.T

# print(adj_mat)

num_trials = 1000

cso_sum = 0
print("==========")
for i in range(num_trials):
    x, y = generate_karger(adj_mat.copy(), i)
    # print(sorted(x))
    # print(sorted(y))
    cso_sum += tanimoto_cso(adj_mat, x, y)

print(f"Karger's: {cso_sum / num_trials}")

cso_sum = 0
print("==========")
for i in range(num_trials):

    x = list(np.random.choice(dummy_size, round(dummy_size * 0.2)))
    y = [i for i in range(dummy_size) if not i in x]    
    # print(sorted(x))
    # print(sorted(y))
    cso_sum += tanimoto_cso(adj_mat, x, y)

print(f"Random: {cso_sum / num_trials}")
```
